```python
# Function to plot the graphs using seaborn with points and custom colors
def plot_graphs_seaborn(data, x_key, y_key, title, x_label, y_label):
    plt.figure(figsize=(10, 6))
    
    # Custom bright colors for the lines
    palette = sns.color_palette("bright")
    
    # Iterate over each model in the data
    for i, (model, values) in enumerate(data.items()):
        df = pd.DataFrame(values)
        # Plot both the line and points
        sns.lineplot(x=df[x_key], y=df[y_key], label=model, color=palette[i % len(palette)], marker="o")
    
    plt.title(title, fontsize=16)
    plt.xlabel(x_label, fontsize=14)
    plt.ylabel(y_label, fontsize=14)
    plt.legend(title="Models", fontsize=12, title_fontsize='13')
    plt.grid(True)
    plt.show()
```

```python
# Plot the graphs using seaborn
plot_graphs_seaborn(data_new, 'Recall', 'Precision', 'Precision-Recall Curve (Updated)', 'Recall', 'Precision')
plot_graphs_seaborn(data_new, 'Confidence', 'Precision', 'Precision-Confidence Curve (Updated)', 'Confidence', 'Precision')
plot_graphs_seaborn(data_new, 'Confidence', 'Recall', 'Recall-Confidence Curve (Updated)', 'Confidence', 'Recall')
```
